```python
from apyori import apriori
```

```python
frequentItemsets = list(apriori(transactions,min_support = 0.075)) # เจอแค่ 1-itemset
frequentItemsets
```

```python
# ลด support เพื่อหา 2-itemset ที่น่าตื่นเต้นกว่า
frequentItemsets = list(apriori(transactions,min_support = 0.015))
frequentItemsets
```

#quiz 10 march
สร้างตัวแปรมา 2 ตัว


*   เก็บ 2-itemset
*   เก็บ 3-itemset



```python
two_itemset = [] # สร้าง list เปล่าไว้เก็บข้อมูล
for itemset in frequentItemsets:
    if len(itemset.items) == 2:
        two_itemset.append(itemset)

# เก็บ 3-itemset
three_itemset = [] # สร้าง list เปล่าไว้เก็บข้อมูล
for itemset in frequentItemsets:
    if len(itemset.items) == 3:
        three_itemset.append(itemset)
```
