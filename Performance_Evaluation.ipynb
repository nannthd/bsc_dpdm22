```python
from sklearn import datasets 
iris_data = datasets.load_iris()
```

```python
X = iris_data['data']
y = iris_data['target']
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score, KFold
```

```python
# Define the decision tree model
model = DecisionTreeClassifier()

# Define the cross-validation method กำหนดวิธีการตรวจสอบความถูกต้องโดยใช้ 10-fold cross-validation
kf = KFold(n_splits=10, shuffle=True, random_state=42)

# Perform k-fold cross-validation and return the accuracy scores
scores = cross_val_score(model, X, y, cv=kf)

# Print the mean accuracy score and standard deviation
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

```python
# Define the decision tree model
model = DecisionTreeClassifier(max_depth=2)

# Define the cross-validation method
kf = KFold(n_splits=10, shuffle=True, random_state=42)

# Perform k-fold cross-validation and return the accuracy scores
scores = cross_val_score(model, X, y, cv=kf)

# Print the mean accuracy score and standard deviation
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

```python
from sklearn.model_selection import LeaveOneOut
```

```python
# Define the decision tree model
dt = DecisionTreeClassifier()

# Define the cross-validation method (leave-one-out) กำหนดวิธีการตรวจสอบความถูกต้องโดยใช้ leave-one-out
loo = LeaveOneOut()

# Perform leave-one-out cross-validation and return the accuracy scores
scores = cross_val_score(dt, X, y, cv=loo)

# Print the mean accuracy score and standard deviation
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```
